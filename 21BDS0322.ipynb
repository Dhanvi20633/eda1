```python
#handling missing values

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/mroz.csv"
df = pd.read_csv(url)

# Drop rows where 'wage' is missing
df = df.dropna(subset=['wage'])

# Select numeric features and drop target from feature set
X = df.drop(columns=['wage']).select_dtypes(include='number')

# Drop rows with any remaining NaNs in features
X = X.dropna()

# Align y with X index
y = df.loc[X.index, 'wage']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print(" Model Evaluation")

print("R² Score:", r2_score(y_test, y_pred))

```
